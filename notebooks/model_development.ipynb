```python
import pandas as pd
```

```python
# Veriyi oku 
data = pd.read_csv('runs/detect/train3/results.csv')
data.head()
```

```python
# Verilerin box, cls, dfl loss degerlerini integer location olarakak degerlerini al 
box_loss = data.iloc[:,2:3].values
cls_loss = data.iloc[:,3:4].values
dfl_loss = data.iloc[:,4:5].values
```

```python
from matplotlib import pyplot as plt
plt.plot(box_loss,label='box loss')
plt.title('Bounding Box Regression Loss', fontsize=20, fontweight='bold')
plt.xlabel('Epouch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(cls_loss,label='classification loss')
plt.title('Classification Loss', fontsize=20, fontweight='bold')
plt.xlabel('Epouch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(dfl_loss,label='distribution focal loss')
plt.title('Distribution Focal Loss', fontsize=20, fontweight='bold')
plt.xlabel('Epouch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
